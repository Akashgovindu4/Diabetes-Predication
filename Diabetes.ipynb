Data Cleaning

```python
df.isnull().any()
```

```python
df.isnull().sum()
```

Converting Zero values to numerical values

```python
columns_with_zeros=['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI']
df[columns_with_zeros]=df[columns_with_zeros].replace(0,np.nan)
```

```python
df.isnull().sum()
```

```python
df[columns_with_zeros]=df[columns_with_zeros].fillna(df.median())
```

```python
df.head()
```

```python
df.isnull().sum()
```

Converting Outcome to category

```python
df['Outcome']=df['Outcome'].astype('category')
```

```python
df.dtypes
```

Outlier Analysis

```python
plt.figure(figsize=(20,10))
sns.boxplot(data=df)
```

```python
q1=df[['SkinThickness','Insulin','BMI']].quantile(0.25)
q3=df[['SkinThickness','Insulin','BMI']].quantile(0.75)
IQR=q3-q1
df=df[((df[['SkinThickness','Insulin','BMI']]<(q1-1.5*IQR)) | (df[['SkinThickness','Insulin','BMI']]>(q3+1.5*IQR))).any(axis=1)]
```

```python
sns.boxplot(data=df)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled_features=['Glucose','BloodPressure','SkinThickness','Insulin','BMI','Age']
df[scaled_features]=scaler.fit_transform(df[scaled_features])
```

```python
sns.heatmap(df.corr(),annot=True)
```

```python
x=df.drop('Outcome',axis=1)
y=df['Outcome']
```

Train_test_split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report,roc_curve,auc
clf_lrs=LogisticRegression()
clf_lrs.fit(x_train,y_train)
```

```python
confusion_matrix(clf_lrs.predict(x_test),y_test)
```

```python
m1=accuracy_score(clf_lrs.predict(x_test),y_test)
m1
```

```python
print(classification_report(clf_lrs.predict(x_test),y_test),end='\n')
```

```python
r1=roc_auc_score(y_test,clf_lrs.predict_proba(x_test)[:,1])
r1
```

Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
clf_tree=DecisionTreeClassifier()
clf_tree.fit(x_train,y_train)
```

```python
y_pred=clf_tree.predict(x_test)
confusion_matrix(y_pred,y_test)
```

```python
m2=accuracy_score(y_pred,y_test)
m2
```

```python
print(classification_report(y_pred, y_test),end="\n")
```

```python
r2=roc_auc_score(y_test,clf_tree.predict_proba(x_test)[:,1])
r2
```

```python
models={
    'LR': LogisticRegression(),
    'DT':DecisionTreeClassifier(),
    'RF': RandomForestClassifier(n_estimators=1000,n_jobs=-1,random_state=42),
    'GRB':GradientBoostingClassifier(),
    'SVM': SVC(kernel='rbf',C=1,gamma='scale',probability=True)
}
roc_auc_scores={}
for name, model in models.items():
    model.fit(x_train,y_train)
    y_probs=model.predict_proba(x_test)[:,1]
    roc_auc=roc_auc_score(y_test,y_probs)
    roc_auc_scores[name]=roc_auc
    fpr,tpr,_=roc_curve(y_test,y_probs)
    plt.plot(fpr,tpr,lw=2,label=f'{name}(AUC={roc_auc:.2f})')
plt.plot([0,1],[0,1],'r--',label='Random Guess')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves for five Models')
plt.legend()
plt.show()
```
